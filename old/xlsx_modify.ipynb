```python
from openpyxl import Workbook
import os
```

```python
wb = openpyxl.load_workbook(os.path.join("ori", "new.xlsx"))
wb
```

```python
ws = wb.active
ws
```

```python
column_values = [ws.cell(row=row, column=3).value for row in range(1, ws.max_row+1)]
len(column_values)
```

```python
unique_values = list(set(column_values))
len(unique_values)
```

```python
# remove duplicates
new_sheet = wb.create_sheet()
for value in unique_values:
    for row in range(1, ws.max_row+1):
        if ws.cell(row=row, column=3).value == value:
            new_sheet.append([ws.cell(row=row, column=col).value for col in range(1, ws.max_column+1)])
            break
wb.remove(ws)
ws=wb.active
ws
```

```python
# sorted by question and category
data = [row for row in ws.iter_rows(values_only=True)]
data.sort(key=lambda x: tuple(x[column-1] for column in [2,3]))
len(data)
new_sheet = wb.create_sheet()
for row_data in data:
    new_sheet.append(row_data)
wb.remove(ws)
ws = wb.active
ws.delete_cols(1)
```

```python
# split by category
```

```python
wb1 = Workbook()
ws1 = wb1.active
wb2 = Workbook()
ws2 = wb2.active
wb3 = Workbook()
ws3 = wb3.active
ws3
```

```python
for row in ws.values:
    print(row)
    row_category = row[0]
    if "判断题" in row_category:
        ws1.append(row)
    elif "单选题" in row_category:
        ws2.append(row)
    elif "多选题" in row_category:
        ws3.append(row)
```

```python
# define styles
from openpyxl.styles import PatternFill, Color
from openpyxl.styles import Font, Alignment, Border, Side
green_fill = PatternFill(start_color="00FF00", end_color="00FF00", fill_type="solid")
thin = Side(border_style="thin", color="000000")
```

```python
# judge
for row_num in range(1, ws1.max_row+1):
    # modify cell bgcolor
    row_data = [ws1.cell(row=row_num, column=col).value for col in range(1, ws1.max_column+1)]
    if row_data[4] == "对":
        ws1.cell(row=row_num, column=3).fill = green_fill
    else:
        ws1.cell(row=row_num, column=4).fill = green_fill
    ws1.cell(row=row_num, column=5).value = None
    ws1.cell(row=row_num, column=2).alignment = Alignment(wrap_text=True)
```

```python
for row in ws1.rows:
    for cell in row:
        cell.alignment = Alignment(wrap_text=True)
        cell.font = Font(size=16)
        cell.border = Border(top=thin, left=thin, right=thin, bottom=thin)
for row in ws2.rows:
    for cell in row:
        cell.alignment = Alignment(wrap_text=True)
        cell.font = Font(size=16)
        cell.border = Border(top=thin, left=thin, right=thin, bottom=thin)
for row in ws3.rows:
    for cell in row:
        cell.alignment = Alignment(wrap_text=True)
        cell.font = Font(size=16)
        cell.border = Border(top=thin, left=thin, right=thin, bottom=thin)
```

```python
# sigle option
for row_num in range(1, ws2.max_row+1):
    row_data = [ws2.cell(row=row_num, column=col).value for col in range(1, ws2.max_column+1)]
    patt = {"A": 3, "B":4, "C":5, "D":6, "E":7, "F":8}
    ws2.cell(row=row_num, column=patt[row_data[6]]).fill = green_fill
    ws2.cell(row=row_num, column=7).value = None
    ws2.cell(row=row_num, column=2).alignment = Alignment(wrap_text=True)
```

```python
# multi options
for row_num in range(1, ws3.max_row+1):
    row_data = [ws3.cell(row=row_num, column=col).value for col in range(1, ws3.max_column+1)]
    i = -1
    while row_data[i] in [None, ""]:
        i-=1
    patt = {"A": 3, "B":4, "C":5, "D":6, "E":7, "F":8}
    for ch in row_data[i]:
        ws3.cell(row=row_num, column=patt[ch]).fill = green_fill
    ws3.cell(row=row_num, column=(ws3.max_column+i+1)).value = None
```
